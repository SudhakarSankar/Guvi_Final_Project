```python

image_path = "C:/Sudhakar/Projects/Guvi Final Project/Image_Processing/Dataset and Document/Text_Image.jpg"

# Load the image
img = cv2.imread(image_path)

# Check if the image is loaded successfully
if img is None:
    print("❌ Error: Image not found. Check the path.")
    exit()

# Function to display images with titles
def show_image(title, image, cmap="gray"):
    plt.figure(figsize=(6, 4))
    plt.imshow(image, cmap=cmap if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image("Grayscale Image", gray)
```

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
show_image("Blurred Image", blurred)
```

```python
edges = cv2.Canny(blurred, 50, 150)
show_image("Canny Edge Detection", edges)
```

```python
kernel = np.ones((5, 5), np.uint8)
show_image("Kernel" , kernel)
```
